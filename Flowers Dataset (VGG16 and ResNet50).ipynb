```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense,GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import VGG16

```

```python
train_dir=('/content/drive/MyDrive/Flower dataset/train')
val_dir=('/content/drive/MyDrive/Flower dataset/test/Validation Test')
```

```python
val_dir
```

```python
train_dir
```

```python
img_size=(224,224)
batch_size=32
```

```python
train_datagen=ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)
```

```python
train_generator=train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)
```

```python
train_generator
```

```python
validation_datagen=ImageDataGenerator(rescale=1./255)
```

```python
validation_generator=validation_datagen.flow_from_directory(
    val_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)
```

```python
base_model=VGG16(weights='imagenet',include_top=False,input_shape=(224,224,3))
```

```python
base_model
```

```python
x=base_model.output
x=GlobalAveragePooling2D()(x)
x=Dense(128,activation='relu')(x)
predictions=Dense(5,activation='sigmoid')(x)
```

```python
model=Model(inputs=base_model.input,outputs=predictions)
```

```python
print(len(train_generator))
```

```python
print(len(validation_generator))
```

```python
for layer in base_model.layers:
  layer.trainable=False
```

```python
model.compile(optimizer=Adam(),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples//batch_size,
    epochs=7,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples//batch_size,
)
```

```python
t_a=history.history['accuracy']
v_a=history.history['val_accuracy']
t_l=history.history['loss']
v_l=history.history['val_loss']
```
